````julia
using LinearAlgebra, CSV, DataFrames


# Vector{<:Number} -> Przyjmuje wektor typów które implementują klasę Number
function dot_general_lib(x::Vector{<:Number}, A::Matrix{<:Number}, y::Vector{<:Number})
    dot(x, A, y)
end

function dot_general(x::Vector{<:Number}, A::Matrix{<:Number}, y::Vector{<:Number})
    dot(x, A * y)
end

function measure_time(func::Function, x::Vector{<:Number}, A::Matrix{<:Number}, y::Vector{<:Number})
    # Number of times test should be measured
    size = 10
    results = zeros(Float64, size)

    for i in range(1, step=1, length=size)
        time_elapsed = @elapsed func(x, A, y)
        results[i] = time_elapsed 
    end

    return results

end

min_val = -100
max_val = 100
test_case = [1000, 5000, 10000, 20000, 50000]


results = DataFrame(FunctionName = String[], VectorSize = Int[], ExecutionTime = Float64[])

for n in test_case
    x = rand(min_val:max_val, n)
    A = rand(min_val:max_val, n, n)
    y = rand(min_val:max_val, n)

    d_gen_out = measure_time(dot_general, x, A, y)
    d_gen_lib_out = measure_time(dot_general_lib, x, A, y)

    append!(results, DataFrame(FunctionName = "dot_general", VectorSize = n, ExecutionTime = d_gen_out))
    append!(results, DataFrame(FunctionName = "dot_general_lib", VectorSize = n, ExecutionTime = d_gen_lib_out))
end

CSV.write("execution_times.csv", results)
````

````julia
using CSV, DataFrames, Plots

input="execution_times.csv"
df=CSV.read(input, delim=",",DataFrame)

grouped = groupby(df, [:FunctionName, :VectorSize])
agg_df = combine(grouped, [:ExecutionTime] => mean => :MeanTime, 
                           [:ExecutionTime] => std => :StdDev,)

function create_plots(agg_df)
        # Plot for dot_general
        plot1 = plot(
                # Filter only rows that have "dot_general" :FunctionName
                filter(row -> row.FunctionName == "dot_general", agg_df).VectorSize, 
                filter(row -> row.FunctionName == "dot_general", agg_df).MeanTime,
                yerror = filter(row -> row.FunctionName == "dot_general", agg_df).StdDev,
                label = "Mean Time",
                title = "dot_general Execution Time",
                xlabel = "Vector Size",
                ylabel = "Execution Time (s)",
                color = :red
        )

        # Plot for dot_general_lib
        plot2 = plot(
                # Filter only rows that have "dot_general_lib" :FunctionName
                filter(row -> row.FunctionName == "dot_general_lib", agg_df).VectorSize, 
                filter(row -> row.FunctionName == "dot_general_lib", agg_df).MeanTime,
                yerror = filter(row -> row.FunctionName == "dot_general_lib", agg_df).StdDev,
                label = "Mean Time",
                title = "dot_general_lib Execution Time",
                xlabel = "Vector Size",
                ylabel = "Execution Time (s)",
                color = :blue
        )

        # (2, 1) -> 2 rows in 1 column
        plot(plot1, plot2, layout = (2, 1), size = (800, 600))
end

create_plots(agg_df)
````
